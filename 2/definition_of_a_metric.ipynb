Puis on créé une matrice avec comme valeur les similitudes entre chaque valeurs: (de 1 à 0)

job: Du plus manuel au moins manuel avec: 
<br>fireman = 1
<br>doctor = 0.5
<br>engineer = 0.5
<br>teacher = 0.25
<br>painter = 0.25
<br>designer = 0
<br>developper = 0


```python
df_job = pd.DataFrame(df['job'].unique(), columns=["job"])
df_job.loc[df_job['job'] == "fireman", "value"] = 1
df_job.loc[df_job['job'] == "doctor", "value"] = 0.5
df_job.loc[df_job['job'] == "engineer", "value"] = 0.5
df_job.loc[df_job['job'] == "teacher", "value"] = 0.25
df_job.loc[df_job['job'] == "painter", "value"] = 0.25
df_job.loc[df_job['job'] == "designer", "value"] = 0
df_job.loc[df_job['job'] == "developper", "value"] = 0

df_job_arr = df_job['value'].values - df_job['value'].values[:, None]
job_matrix = pd.concat((df_job['job'], pd.DataFrame(df_job_arr, columns=df_job['job'])), axis=1)
job_matrix
```

city: Du Nord au Sud avec: 
<br>lille = 1
<br>paris = 0.75
<br>toulouse = 0.5
<br>marseille = 0.25
<br>madrid = 0

```python
df_city = pd.DataFrame(df['city'].unique(), columns=["city"])
df_city.loc[df_city['city'] == "lille", "value"] = 1
df_city.loc[df_city['city'] == "paris", "value"] = 0.75
df_city.loc[df_city['city'] == "toulouse", "value"] = 0.5
df_city.loc[df_city['city'] == "marseille", "value"] = 0.25
df_city.loc[df_city['city'] == "madrid", "value"] = 0

df_city_arr = df_city['value'].values - df_city['value'].values[:, None]
city_matrix = pd.concat((df_city['city'], pd.DataFrame(df_city_arr, columns=df_city['city'])), axis=1)
city_matrix
```

favorite music style: Du classique au technical death metal avec: 
<br>classical = 1
<br>jazz = 0.8

<br>rap = 0.5
<br>other = 0.5

```python
df_favorite_music_style = pd.DataFrame(df['favorite music style'].unique(), columns=["favorite music style"])
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "classical", "value"] = 1
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "jazz", "value"] = 0.8
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "hiphop", "value"] = 0.6
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "trap", "value"] = 0.6
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "rap", "value"] = 0.5
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "other", "value"] = 0.5
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "rock", "value"] = 0.3
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "metal", "value"] = 0.1
df_favorite_music_style.loc[df_favorite_music_style['favorite music style'] == "technical death metal", "value"] = 0

df_favorite_music_style_arr = df_favorite_music_style['value'].values - df_favorite_music_style['value'].values[:, None]
favorite_music_style_matrix = pd.concat((df_favorite_music_style['favorite music style'], pd.DataFrame(df_favorite_music_style_arr, columns=df_favorite_music_style['favorite music style'])), axis=1)
favorite_music_style_matrix
```

On definie une fonction pour simplifier le calcule de la matrice de similitude: 
<br> job, city & favorite music style sont emplifié du a leur importance par rapport à l'age et à la taille

```python
def compute_dissimilarity(id_1, id_2):
    id_1_age = df.loc[id_1]['age']
    id_2_age = df.loc[id_2]['age']

    id_1_heigth = df.loc[id_1]['height']
    id_2_heigth = df.loc[id_2]['height']

    id_1_job = df.loc[id_1]['job']
    id_2_job = df.loc[id_2]['job']
    dissimilarity_job = abs(job_matrix[id_1_job].loc[job_matrix["job"] == id_2_job].values[0])

    id_1_city = df.loc[id_1]['city']
    id_2_city = df.loc[id_2]['city']
    dissimilarity_city = abs(city_matrix[id_1_city].loc[city_matrix["city"] == id_2_city].values[0])

    id_1_favorite_music_style = df.loc[id_1]['favorite music style']
    id_2_favorite_music_style = df.loc[id_2]['favorite music style']
    dissimilarity_favorite_music_style = abs(favorite_music_style_matrix[id_1_favorite_music_style].loc[favorite_music_style_matrix["favorite music style"] == id_2_favorite_music_style].values[0])


    dissimilarity = math.sqrt(
        (id_1_age - id_2_age) ** 2 +
        (id_1_heigth - id_2_heigth) ** 2 +
        10 * dissimilarity_job +
        5 * dissimilarity_city +
        20 * dissimilarity_favorite_music_style
    )

    return(dissimilarity)
```

On calcule la matrice de désimiliralité pour tout les index du dataframe:

```python
nb_ids = len(df.index)
dissimilarity_matrix = np.zeros((nb_ids, nb_ids))

for id_1 in range(nb_ids):
    for id_2 in range(nb_ids):
        dissimilarity = compute_dissimilarity(id_1, id_2)
        dissimilarity_matrix[id_1, id_2] = dissimilarity
dissimilarity_matrix
```

On sauvegarde la matrice en .npy:

```python
np.save('./dissimilarity_matrix.npy', dissimilarity_matrix)
```
