```python
import numpy as np
import pandas as pd
```

Creation du dataset:

```python
df = pd.DataFrame({
    "int": np.random.randint(0, 100, 300), 
    "float": np.random.uniform(0, 100, 300),
    "mean=2.5": np.random.normal(2.5, 5, 300),
    "binomial": np.random.binomial(10, 0.5, 300),
    "neg_binomial": np.random.negative_binomial(10, 0.5, 300),
    "poisson": np.random.poisson(10, 300),
})
df
```

On regarde la correlation:

```python
corr_matrix = df.corr()
corr_matrix
```

On export le dataset créé en csv:

```python
df.to_csv("./artificial_dataset.csv")
```
